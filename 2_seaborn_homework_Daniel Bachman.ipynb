```python
# Import the Palmer penguins dataset into a pandas DataFrame
df = sns.load_dataset('penguins')

# Convert to category data type
df[['species', 'island', 'sex']] = df[['species', 'island', 'sex']].astype('category')

# Drop rows with missing values
df = df.dropna().reset_index(drop=True)

# Look at the first 5 rows of the data
df.head()
```

# Line Plot

Here, you should use the seaborn library to create a seaborn line chart, with the `"bill_length_mm"` column along the x-axis and the `"bill_depth_mm"` column along the y-axis. Remember to add a title to your chart and label the x- and y-axis!

Seaborn `lineplot` [documentation](https://seaborn.pydata.org/generated/seaborn.lineplot.html)

```python
# Create a line plot using seaborn
plt.figure(figsize=(10,5))

ax = sns.lineplot(data=df, x='bill_length_mm', y='bill_depth_mm', ci=None)

plt.title('Bill Length & Depth Relationship',size=15)
plt.xlabel('Bill_Length_mm',size=15)
plt.ylabel('Bill_Depth_mm',size=15);
```

# Count Plot

Seaborn's `countplot()` method can be thought of as a histogram across a categorical field, instead of quantitative, variable. 

Here, you should use the `countplot()` method to find the answer to the following question.

**Question:** Which island has the most penguins?  
**Answer**: Biscoe has the most penguins

Seaborn `countplot` [documentation](https://seaborn.pydata.org/generated/seaborn.countplot.html#seaborn.countplot)

```python
# Create a count plot using the "island" field
ax = sns.countplot(x='island', data=df)
```

To answer the next question, this time you should take advantage of the `countplot()` function's optional `hue` parameter. In addition, you will also need to:
- Plot the bars horizontally
- Add a title and label the x- and y-axis
- Select one seaborn's optional color palettes: `"deep"`, `"muted"`, `"pastel"`, `"bright"`, `"dark"`, `"colorblind"`

**Question:** On Biscoe island, which sex is predominate?  
**Answer:** Male is the predominate sex on Biscoe Island

```python
# Create a countplot
ax = sns.countplot(y='island', data=df,hue='sex',palette='colorblind')

plt.title('Penguin Gender Breakdown by Island',size=12)
plt.xlabel('Count',size=12)
plt.ylabel('Island',size=12);
```

# Cat Plot

Seaborn's `catplot()` is a figure-level interface that provides access to several axes-level functions that we can use to show the relationship between a numerical variable and one or more categorical variables in our dataset.

To answer the following question, you should use `catplot` to create a count plot. If you forget how to do this, be sure to check out the documentation or the inclass demo.)

**Question:** Do any of the islands only have one species of penguin?   
**Answer:** Yes, Torgersen has only one species of penguin (The Adelie)

Seaborn `catplot` [documentation](https://seaborn.pydata.org/generated/seaborn.catplot.html)

```python
# Create a catplot
g = sns.catplot(data=df, x='species', kind='count', hue='sex', col='island')
```

# Histogram

Similar to `catplot`, seaborn's `displot()` method is a figure-level interface to several axes-level functions such as `histplot()` and `kdeplot()`. 

Here, you should use `displot()` to visualize the distribution of the `"flipper_length_mm"` variable. Then briefly describe in no more than one or two sentences whether or not you think the feature is normally distributed. 

**Answer:** Generally speaking the distribution does have a bell curve to it, however it is skewed to the right. From this standpint, we would not say it is normally distributed.

Seaborn `displot` [documentation](https://seaborn.pydata.org/generated/seaborn.displot.html)

```python
# Create a displot
g = sns.displot(data=df, x='flipper_length_mm')
```

Here, you are interested in learning more about the distribution of the penguins' flipper lengths. Using `displot()` again, this time pass in the `"species"` variable to the `hue` parameter so that it will draw a separate histogram for each of its unique values and distinguish them by color.

```python
# Create a displot
g = sns.displot(data=df, x='flipper_length_mm', hue='species')
```

That last plot was a little difficult to read! However, since `displot()` is a figure-level function, we can use it to draw each individual distribution in a separate subplot for each unique species. If you don't remember which optional parameter to use, just check out the `displot()` documentation!

Seaborn `displot` [documentation](https://seaborn.pydata.org/generated/seaborn.displot.html)

```python
# Create a displot
g = sns.displot(data=df, x='flipper_length_mm', col='species')
```
